```python
import json
import torch
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    BitsAndBytesConfig,
    TrainingArguments,
    Trainer,
    DataCollatorForLanguageModeling
)
from peft import (
    LoraConfig,
    get_peft_model,
    prepare_model_for_kbit_training
)

# ==========================
# 1. LOAD DATA
# ==========================

with open("master_training_320_stories_file.txt", "r", encoding="utf-8") as f:
    train_data = json.load(f)

with open("Validation_Stories.txt", "r", encoding="utf-8") as f:
    val_data = json.load(f)

print("Train stories:", len(train_data))
print("Validation stories:", len(val_data))

# ==========================
# 2. FORMAT DATA
# ==========================

def format_example(example):
    prompt = f"""
You are a narrative decomposition model.

Convert story into structured JSON scenes.

Story:
{example['input']}

Output:
"""

    response = json.dumps(example["output"], ensure_ascii=False)

    return {
        "text": prompt + response
    }

train_formatted = [format_example(x) for x in train_data]
val_formatted = [format_example(x) for x in val_data]

train_dataset = Dataset.from_list(train_formatted)
val_dataset = Dataset.from_list(val_formatted)

# ==========================
# 3. LOAD TOKENIZER
# ==========================

from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
import torch

model_name = "meta-llama/Meta-Llama-3-8B-Instruct"

# tokenizer
tokenizer = AutoTokenizer.from_pretrained(
    model_name,
    token="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"
)

tokenizer.pad_token = tokenizer.eos_token

# 4-bit quantization (QLoRA)
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.float16
)

# model
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map="auto",
    token="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"
)

model.config.pad_token_id = tokenizer.pad_token_id

print("Model loaded successfully")

# ==========================
# 4. QLORA CONFIG
# ==========================

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True
)

# ==========================
# 5. LOAD MODEL
# ==========================

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map="auto",
    token="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"
)

model = prepare_model_for_kbit_training(model)

# ==========================
# 6. LORA ADAPTER
# ==========================

lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=[
        "q_proj",
        "k_proj",
        "v_proj",
        "o_proj"
    ],
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)

model = get_peft_model(model, lora_config)

model.print_trainable_parameters()

# ==========================
# 7. TOKENIZATION
# ==========================

def tokenize_function(examples):
    return tokenizer(
        examples["text"],
        truncation=True,
        padding="max_length",
        max_length=2048
    )

train_dataset = train_dataset.map(
    tokenize_function,
    batched=True
)

val_dataset = val_dataset.map(
    tokenize_function,
    batched=True
)

# ==========================
# 8. DATA COLLATOR
# ==========================

data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False
)

# ==========================
# 9. TRAINING ARGUMENTS
# ==========================

training_args = TrainingArguments(
    output_dir="./llama_checkpoints",

    per_device_train_batch_size=1,
    gradient_accumulation_steps=8,   # reduced memory pressure

    num_train_epochs=3,              # enough for 320 samples

    learning_rate=2e-4,

    logging_steps=10,

    save_strategy="epoch",

    # REMOVE evaluation during training
    evaluation_strategy="no",

    save_total_limit=3,

    fp16=True,

    optim="paged_adamw_8bit",   # huge memory saver

    gradient_checkpointing=True,

    report_to="none"
)

# ==========================
# 10. TRAINER
# ==========================

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    data_collator=data_collator
)

# ==========================
# 11. TRAIN
# ==========================

trainer.train()

# ==========================
# 12. SAVE MODEL
# ==========================

model.save_pretrained("./final_llama_model")
tokenizer.save_pretrained("./final_llama_model")

print("Training complete")
```

```python
import json
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import PeftModel

# -----------------------------
# 1. Load validation data
# -----------------------------
print("Loading validation file...")

with open("Validation_Stories.txt", "r") as f:
    val_data = json.load(f)

print(f"Total validation samples: {len(val_data)}")


# -----------------------------
# 2. Load fine-tuned model
# -----------------------------
print("Loading fine-tuned model...")

model_path = "./final_llama_model"

tokenizer = AutoTokenizer.from_pretrained(model_path)

model = AutoModelForCausalLM.from_pretrained(
    model_path,
    torch_dtype=torch.float16,
    device_map="auto"
)

print("Model loaded successfully!")


# -----------------------------
# 3. Generate predictions
# -----------------------------
predictions = []

for i, sample in enumerate(val_data):

    print(f"Processing story {i+1}/{len(val_data)}")

    # adjust key names if needed
    input_story = sample["input"]

    prompt = f"""
Convert the following story into structured scene JSON.

Story:
{input_story}

Output:
"""

    inputs = tokenizer(
        prompt,
        return_tensors="pt",
        truncation=True,
        max_length=512
    ).to("cuda")

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=300,
            temperature=0.7,
            do_sample=False
        )

    generated_text = tokenizer.decode(
        outputs[0],
        skip_special_tokens=True
    )

    predictions.append({
        "input_story": input_story,
        "ground_truth": sample["output"],
        "prediction": generated_text
    })


# -----------------------------
# 4. Save predictions
# -----------------------------
output_file = "validation_predictions.json"

with open(output_file, "w") as f:
    json.dump(predictions, f, indent=4)

print(f"Validation predictions saved to {output_file}")
```

```python
import re
import json
import torch
from tqdm import tqdm

# -----------------------------------
# 1. Load validation stories file
# -----------------------------------
validation_file = "/teamspace/studios/this_studio/ComfyUI/tests/all_validation_Stories.txt"

with open(validation_file, "r", encoding="utf-8") as f:
    raw_text = f.read()

# Split using Story 1, Story 2... pattern
validation_stories = re.split(r"Story\s+\d+:", raw_text)

# Remove empty splits
validation_stories = [s.strip() for s in validation_stories if s.strip()]

print("Total validation stories loaded:", len(validation_stories))
print("\nFirst story preview:\n")
print(validation_stories[0][:500])

# -----------------------------------
# 2. Generate predictions
# -----------------------------------
all_predictions = []

for i, story in enumerate(tqdm(validation_stories)):

    prompt = f"""
Perform narrative decomposition of the following story.

Rules:
- Break story into logical scenes
- Maintain chronological order
- Preserve transformation arc
- Output valid JSON only

Story:
{story}

JSON Output:
"""

    inputs = tokenizer(
        prompt,
        return_tensors="pt",
        truncation=True,
        max_length=700
    ).to("cuda")

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=700,
            temperature=0.2,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id
        )

    prediction = tokenizer.decode(
        outputs[0], 
        skip_special_tokens=True
    )

    all_predictions.append({
        "story_id": i + 1,
        "input_story": story,
        "predicted_output": prediction
    })

    print(f"Finished story {i+1}/40")

# -----------------------------------
# 3. Save predictions
# -----------------------------------
with open("validation_predictions.json", "w", encoding="utf-8") as f:
    json.dump(all_predictions, f, indent=4, ensure_ascii=False)

print("Validation predictions saved successfully!")
```

```python
import json
import re
import torch
from tqdm import tqdm

# -----------------------------------
# 1. Load test stories file
# -----------------------------------
test_file = "/teamspace/studios/this_studio/ComfyUI/test_stories.txt"

with open(test_file, "r", encoding="utf-8") as f:
    raw_text = f.read()

print("Loaded test stories file successfully")


# -----------------------------------
# 2. Split stories correctly
# -----------------------------------
stories = re.split(r"Story\s+\d+:", raw_text)
stories = [s.strip() for s in stories if s.strip()]

print(f"Total test stories found: {len(stories)}")


# -----------------------------------
# 3. Generate decompositions
# -----------------------------------
all_test_predictions = []

for i, story in enumerate(tqdm(stories)):

    prompt = f"""
Perform narrative decomposition of the following story.

Return ONLY valid JSON array.
Do not repeat the prompt.
Do not explain anything.
Maintain chronological scene flow.

Story:
{story}

JSON:
"""

    inputs = tokenizer(
        prompt,
        return_tensors="pt",
        truncation=True,
        max_length=700
    ).to("cuda")

    input_length = inputs["input_ids"].shape[1]

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=700,
            temperature=0.2,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id
        )

    # ONLY decode generated tokens
    generated_tokens = outputs[0][input_length:]

    prediction = tokenizer.decode(
        generated_tokens,
        skip_special_tokens=True
    )

    all_test_predictions.append({
        "story_id": i + 1,
        "input_story": story,
        "predicted_output": prediction
    })


# -----------------------------------
# 4. Save final outputs
# -----------------------------------
output_file = "final_test_predictions.json"

with open(output_file, "w", encoding="utf-8") as f:
    json.dump(all_test_predictions, f, indent=4, ensure_ascii=False)

print(f"\nDone! Final predictions saved in: {output_file}")
```
